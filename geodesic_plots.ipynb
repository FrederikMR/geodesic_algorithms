# Plotting Geodesic Estimates

```python
#Modules
import jax.numpy as jnp

#plotting
import matplotlib.pyplot as plt

#Own modules
from load_manifold import load_manifold
from jaxman.optimization import GeodesicControl, ScipyMinimization, ODEMinimization, GradientDescent
```

```python
#hyper parameters
N = 20
T = 1.0
tau = 1.0
max_iter = 1000
tol = 1e-3
```

## Plotting Geodesics on Simple Manifolds

### Paraboloid

```python
x0,xT, M = load_manifold("Paraboloid", dim = 2)
x0 = jnp.array([jnp.cos(2*jnp.pi/4), jnp.sin(2*jnp.pi/4)])
xT = jnp.array([-1.0, 0.0])
```

```python
Geo = GradientDescent(M,
                      tau=tau,
                      T=T,
                      N=N,
                      tol=tol,
                      max_iter=max_iter,
                      optimizer="Adam"
                      )
gradient_descent = Geo(x0,xT, "While")
gradient_energy = M.energy(gradient_descent[0])
```

```python
Geo = ScipyMinimization(M,
                        T=T,
                        N=N,
                        tol=tol,
                        max_iter=max_iter,
                        method="BFGS",
                        )
scipyminimization = Geo(x0,xT)
scipy_energy = M.energy(scipyminimization[0])
```

```python
Geo = GeodesicControl(M=M,
                      tau=tau,
                      T=1.0,
                      N=N,
                      tol=tol,
                      max_iter=max_iter,
                      inv_method="Naive"
                      )
controlopt = Geo(x0,xT, "While")
control_energy = M.energy(controlopt[0])
```

```python
x1, x2 = jnp.linspace(-2.0,2.0,100), jnp.linspace(-2.0,2.0,100)
X1,X2 = jnp.meshgrid(x1,x2)
X = jnp.concatenate((X1.reshape(1,100,100), X2.reshape(1,100,100), 
                     X1.reshape(1,100,100)**2+X2.reshape(1,100,100)**2), axis=0)

fig = plt.figure(figsize=(20,10))

ax = fig.add_subplot(121)
ax.plot(controlopt[0][:,0], controlopt[0][:,1], color='red', label='Control')
ax.plot(scipyminimization[0][:,0], scipyminimization[0][:,1], color='blue', label='BFGS')
ax.plot(gradient_descent[0][:,0], gradient_descent[0][:,1], color='purple', label='Adam')

ax = fig.add_subplot(122, projection='3d')
ax.plot_surface(X[0], X[1], X[2], alpha=0.2, color='cyan')
ax.plot(controlopt[0][:,0], controlopt[0][:,1], 
        controlopt[0][:,0]**2+controlopt[0][:,1]**2, color='red', label='Control')
ax.plot(scipyminimization[0][:,0], scipyminimization[0][:,1], 
        scipyminimization[0][:,0]**2+scipyminimization[0][:,1]**2, color='blue', label='BFGS')
ax.plot(gradient_descent[0][:,0], gradient_descent[0][:,1], 
        gradient_descent[0][:,0]**2+gradient_descent[0][:,1]**2, color='purple', label='Adam')
```
