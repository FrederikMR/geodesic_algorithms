# Geodesics on Learned Manifolds

```python
#Modules
import jax.numpy as jnp
import jax.random as jrandom
from jax import vmap

import haiku as hk

import torchvision.utils as vutils

#numpy
import numpy as np

from abc import ABC

import torch

import matplotlib.pyplot as plt

from geodesics import GradientDescent, JAXOptimization, ScipyOptimization, GC_LineSearch

from vae.model_loader import mnist_generator, svhn_generator, celeba_generator, load_model

from vae.models.train.mnist import Encoder as mnist_encoder
from vae.models.train.mnist import Decoder as mnist_decoder
from vae.models.train.mnist import VAE as mnist_vae

from vae.models.train.svhn import Encoder as svhn_encoder
from vae.models.train.svhn import Decoder as svhn_decoder
from vae.models.train.svhn import VAE as svhn_vae

from vae.models.train.celeba import Encoder as celeba_encoder
from vae.models.train.celeba import Decoder as celeba_decoder
from vae.models.train.celeba import VAE as celeba_vae
```

```python
seed = 2712
model_path = 'models/'
```

## VAE MNIST

### Loading Data

```python
mnist_dataloader = mnist_generator(seed=2712,
                                   batch_size=64,
                                   split='train[:80%]')
```

### Loading Model

```python
@hk.transform_with_state
def mnist_tvae(x):


    vae = mnist_vae(
                encoder=mnist_encoder(latent_dim=8),
                decoder=mnist_decoder(),
    )

    return vae(x)

@hk.transform_with_state
def mnist_tencoder(x):


    encoder = mnist_encoder(latent_dim=8)

    return encoder(x)[0]

@hk.transform_with_state
def mnist_tdecoder(x):

    decoder = mnist_decoder(latent_dim=8)

    return decoder(x)
```

```python
mnist_state = load_model(''.join((model_path, 'mnist/')))
```

```python
mnist_encoder_fun = lambda x: mnist_tencoder.apply(mnist_state.params, mnist_state.state_val, mnist_state.rng_key, x)
mnist_decoder_fun = lambda x: mnist_tdecoder.apply(mnist_state.params, mnist_state.state_val, mnist_state.rng_key, x)
mnist_vae_fun = lambda x: mnist_tvae.apply(mnist_state.params, mnist_state.state_val, mnist_state.rng_key, x)
```

```python
mnist_encoder_fun(next(mnist_dataloader).x)
```

```python
mnist_state.params.keys()
```

```python
test, _ = mnist_vae_fun(next(mnist_dataloader).x)
```

```python
mu_zx = test.mu_zx[0]
```

### Plotting Reconstruction

```python
mnist_data = next(mnist_dataloader).x
```

```python
mnist_rec, _ = mnist_vae_fun(mnist_data)
```

```python
plot_val = torch.from_numpy(np.transpose(np.array(mnist_data), (0,3,1,2)))
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.set_axis_off()
ax.set_title("Original Images")
ax.imshow(np.transpose(vutils.make_grid(plot_val, padding=2, normalize=True).cpu(),(1,2,0)))
```

```python
plot_val = torch.from_numpy(np.transpose(np.array(mnist_rec.mu_xz), (0,3,1,2)))
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.set_axis_off()
ax.set_title("Original Images")
ax.imshow(np.transpose(vutils.make_grid(plot_val, padding=2, normalize=True).cpu(),(1,2,0)))
```
